```python
import csv
import matplotlib.pyplot as plt


files = ["results_BETTER/random-few.csv", "results_BETTER_individuals/random-few.csv", "results_BETTER_simplified/random-few.csv", "results_BETTER_simplified_individuals/random-few.csv"]
data = []
for file in files:
    with open(file, 'r') as f:
        reader = csv.reader(f)
        header = next(reader)
        rows = list(reader)
        col_index = header.index("better")
        iter = 0
        file_data = []
        prev_k = -1
        for row in rows:
            k = int(row[0])
            file_data.append(row[col_index])
        data.append(file_data)
```

```python
tags = ["All", "Individuals", "Simplified", "Simplified_Individuals"]

# Plot the results from the 4 CSV files
plt.figure(figsize=(10, 6))
ks = list(range(0,61))


# Plot each file's data
for i, tag in enumerate(tags):
    plt.plot(ks, [float(score) for score in data[i]], marker='o', label=tag)

plt.xlabel('Random Few-shot k')
plt.ylabel('Combined Better F1 Score')
plt.title('Combined Better F1 Score in Few-shot k for Different Languages')
plt.legend()
plt.grid(True)
plt.show()
```
